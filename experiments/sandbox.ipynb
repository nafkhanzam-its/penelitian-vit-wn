```python
#? TRAIN
from sklearn import metrics
from iRF import iRFClassifier
import numpy as np

# Create a Random Forest classifier
model = iRFClassifier()
# Create the random grid
random_grid = {
    "K": np.linspace(1, 10, dtype=int, num=5),
    "B": np.linspace(10, 50, dtype=int, num=10),
    "random_state_classifier": [42],
    "propn_n_samples": [0, .2, .4],
    "bin_class_type": [1],
    "M": np.linspace(0, 24, dtype=int, num=7),
    "max_depth": np.linspace(2, 7, dtype=int, num=6),
    "noisy_split": [False, True],
    "num_splits": [2, 3, 4],
    "signed": [False, True],
    "n_estimators": np.linspace(100, 1000, dtype=int, num=10),
    "n_estimators_bootstrap": np.linspace(5, 50, dtype=int, num=10),
}
# random_grid = {
#     "K": [5],
#     "B": [1],
#     "random_state_classifier": [42],
#     "propn_n_samples": [.2],
#     "bin_class_type": [1],
#     "M": [1],
#     "max_depth": [1],
#     "noisy_split": [False],
#     "num_splits": [2],
#     "signed": [True],
#     "n_estimators_bootstrap": [1],
# }
clf = RandomizedSearchCV(
    model,
    random_grid,
    cv=5,
    random_state=42,
    n_iter=10,
    n_jobs=5,
    verbose=3,
    scoring=lambda est, X_test, y_test: metrics.accuracy_score(y_test, est.predict(X_test)),
    refit=True,
)
search = clf.fit(X, y)
search
```
